```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_data = eng_data[['Outside North Temperature']]
x_data = eng_data[['Living Room Temperature']]

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)

r_score = r2_score(y_test, predicted_values)
round(r_score,2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
eng_data2 = eng_data.drop(columns=['Date and Time','Light Energy','Total Energy Used'])
norm_data = pd.DataFrame(scaler.fit_transform(eng_data2),columns=eng_data2.columns)
target_data = eng_data['Total Energy Used']
x_train, x_test, y_train, y_test = train_test_split(norm_data, target_data, test_size=0.3, random_state=42)
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)

compare = pd.DataFrame({'Actual values':y_test,'Predicted':predicted_values})
compare
```

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
import numpy as np
rss = np.sum(np.square(y_test-predicted_values))
round(rss, 2)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse,3)
```
